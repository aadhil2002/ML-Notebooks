```python
import fitz  # PyMuPDF
import io
import os
from langchain.text_splitter import CharacterTextSplitter

# Function to extract text from a PDF file
def extract_text_from_pdf(pdf_stream):
    text_content = ''

    # Read the PDF content from the streaming body
    pdf_bytes = io.BytesIO(pdf_stream.read())

    # Open the PDF document
    with fitz.open(stream=pdf_bytes, filetype="pdf") as pdf_document:
        # Iterate through each page
        for page_num in range(len(pdf_document)):
            page = pdf_document[page_num]
            text_content += page.get_text()

    return text_content

# Function to split the extracted text into chunks
def split_text_into_chunks(text, chunk_size=384, chunk_overlap=0):
    text_splitter = CharacterTextSplitter(chunk_size=chunk_size, chunk_overlap=chunk_overlap)
    texts = text_splitter.split_text(text)
    return texts

# Function to process all PDF files in a folder
def process_pdfs_in_folder(folder_path):
    all_texts = []

    # Iterate through all files in the folder
    for filename in os.listdir(folder_path):
        if filename.lower().endswith('.pdf'):
            file_path = os.path.join(folder_path, filename)
            print(f"Processing file: {file_path}")  # Debug statement
            with open(file_path, 'rb') as pdf_file:
                # Extract and split text
                text_content = extract_text_from_pdf(pdf_file)
                texts = split_text_into_chunks(text_content)
                all_texts.append({
                    'filename': filename,
                    'chunks': texts
                })

    return all_texts

# Usage example
folder_path = 'Functionality1/Functionality1/Civil Law'  # Ensure correct path
all_texts = process_pdfs_in_folder(folder_path)

# Verify the texts
for pdf_data in all_texts:
    print(f"File: {pdf_data['filename']}")
    for i, text in enumerate(pdf_data['chunks']):
        print(f"  Chunk {i+1}:\n{text}\n")
    print(f"Total number of chunks in {pdf_data['filename']}: {len(pdf_data['chunks'])}")
```

```python
import re

def extract_case_details(text):
    # Define regex patterns for each field
    case_id_pattern = r'S.B. Civil Revision Petition No\. (\d+ of \d{4})'
    case_text_pattern = r'(1989\(2\) RLW 1.+)(?=\(a\) Civil Procedure Code)'
    jurisdiction_pattern = r'Rajasthan High Court, [\w\s]+'
    case_type_pattern = r'Civil Revision Petition'
    court_decision_pattern = r'(Revision disposed with directions\.)'
    similarity_score_pattern = r'Similarity Score: (\d+\.?\d*)'  # Assuming this is in the text

    # Extract the required fields using regex patterns
    case_id_match = re.search(case_id_pattern, text)
    case_text_match = re.search(case_text_pattern, text, re.DOTALL)
    jurisdiction_match = re.search(jurisdiction_pattern, text)
    case_type_match = re.search(case_type_pattern, text)
    court_decision_match = re.search(court_decision_pattern, text)
    similarity_score_match = re.search(similarity_score_pattern, text)  # Assuming this is in the text

    # Get the matched text or None if not found
    case_id = case_id_match.group(1) if case_id_match else None
    case_text = case_text_match.group(1) if case_text_match else None
    jurisdiction = jurisdiction_match.group(0) if jurisdiction_match else None
    case_type = case_type_match.group(0) if case_type_match else None
    court_decision = court_decision_match.group(1) if court_decision_match else None
    similarity_score = similarity_score_match.group(1) if similarity_score_match else None

    return {
        "case_id": case_id,
        "case_text": case_text,
        "jurisdiction": jurisdiction,
        "case_type": case_type,
        "court_decision": court_decision,
        "similarity_score": similarity_score,
        "embeddings":text_embedding_vector
    }


# Extract and print case details for each chunk of text
for text in texts:
    case_details = extract_case_details(text)
    print(case_details)
```
